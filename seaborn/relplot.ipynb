```python
# Draw a single facet to use the FacetGrid legend placement
import seaborn as sns
sns.set(style = "ticks")
tips = sns.load_dataset("tips")
tips.head()
```

```python
g = sns.relplot(x = "total_bill" , y = "tip" , hue = "day" , data = tips)
```

```python
g = sns.relplot(x = "total_bill" , y = "tip" , hue = "day" ,kind = "line", data = tips)
```

```python
# Facet on the columns with another variable:
g = sns.relplot(x = "total_bill" ,y = "tip" , hue = "day" , col = "time" , data = tips)
```

```python
# Facet on the columns and rows:
g = sns.relplot(x = "total_bill" ,y = "tip" , hue = "day" , col = "time" , row = "sex", data = tips)
```

```python
# “Wrap” many column facets into multiple rows:
g = sns.relplot(x = "total_bill" , y = "tip" , hue = "time" , col = "day" , col_wrap = 2, data = tips)
```

```python
# Use multiple semantic variables on each facet with specified attributes:
g = sns.relplot(x = "total_bill",y="tip",hue = "time" , size = "size" , palette = ['b','r'],sizes = (0,100), col = "time", data = tips)
```

```python
# Use a different kind of plot:
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
g = sns.relplot(x = "timepoint" , y = "signal" , hue = "event" , style = "event" , col = "region" , kind = "line" , data = fmri)
```

```python
# Change the size of each facet:
g = sns.relplot(x = "timepoint" , y = "signal" , hue = "event" , style = "event" , col = "region", height = 5, aspect = .7, kind = "line", data = fmri)
```
