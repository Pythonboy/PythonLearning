```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "whitegrid")
tips = sns.load_dataset("tips")
ax = sns.stripplot(x = tips["total_bill"])
```

```python
# Group the strips by a categorical variable:
ax = sns.stripplot(x = "day", y = "total_bill", data =tips)
```

```python
# Add jitter to bring out the distribution of values:
ax = sns.stripplot(x = "day", y = "total_bill", jitter = 0.5, data =tips)
```

```python
# Draw horizontal strips:
ax = sns.stripplot(y = "day", x = "total_bill", data =tips)
```

```python
ax = sns.stripplot(y = "day", x = "total_bill", data =tips,orient = "h")
```

```python
# Draw outlines around the points:
ax = sns.stripplot(x="total_bill", y="day", data=tips, jitter=True, linewidth=1)
```

```python
ax = sns.stripplot(x="sex", y="total_bill", hue="day",data = tips, jitter = True)
```

```python
ax = sns.stripplot(x="day", y="total_bill", hue="smoker",data=tips, jitter=True,palette="Set2", dodge=True)
```

```python
ax = sns.stripplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

```python
ax =  sns.stripplot("day", "total_bill", "smoker", data=tips,palette="Set2", size=20, marker="D", edgecolor="gray", alpha=.25)
```

```python
ax = sns.boxplot(x="tip", y="day", data=tips)
ax = sns.stripplot(x="tip", y="day", data=tips,jitter=True, color=".3")
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips,inner=None, color=".8")
ax = sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```
